```python
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# 載入 Iris 資料集
iris = load_iris()
X, y = iris.data, iris.target  # 特徵和目標

# 分割訓練集和測試集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 建立決策樹模型
tree_clf = DecisionTreeClassifier(criterion='entropy', max_depth=3, random_state=0)
tree_clf.fit(X_train, y_train)

# 預測並評估模型
y_pred = tree_clf.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Decision Tree Accuracy: {accuracy:.2f}")

# 繪製決策樹圖
plt.figure(figsize=(12, 8))
plot_tree(tree_clf, feature_names=iris.feature_names, class_names=iris.target_names, filled=True, rounded=True)
plt.title("Decision Tree Diagram")
plt.show()
plt.save
```
